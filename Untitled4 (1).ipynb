```python
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
ans-
A Bokeh server is a Python process that allows you to create and serve interactive Bokeh plots and applications on a web page. With the Bokeh server, you can create interactive plots that can be updated in real-time based on user input or external data sources.

To use the Bokeh server, you first need to create a Bokeh application. A Bokeh application is a Python script that defines a Bokeh plot and any necessary widgets or interactions. Here is an example of a simple Bokeh application:

python
Copy code
from bokeh.plotting import figure, curdoc
from bokeh.models import Slider

# Define the data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title='Interactive plot', x_axis_label='X-axis', y_axis_label='Y-axis')
line = p.line(x, y)

# Define a callback function for the slider
def update(attr, old, new):
    line.data_source.data['y'] = [i * slider.value for i in y]

# Create a slider widget
slider = Slider(start=0.1, end=2, value=1, step=0.1, title='Multiplier')
slider.on_change('value', update)

# Add the slider to the document
curdoc().add_root(p)
curdoc().add_root(slider)
In this example, we create a Bokeh plot with a line glyph and a slider widget. The update function is a callback function that is called whenever the value of the slider changes. The function updates the y-data of the line glyph based on the value of the slider. We then create a slider widget and attach the update function to its value property. Finally, we add the plot and the slider to the document using the curdoc function.

To run the Bokeh application, you can start the Bokeh server using the command bokeh serve followed by the path to the Python script. For example:

Copy code
bokeh serve myapp.py
This will start the Bokeh server and serve the Bokeh application on a web page. You can access the web page in a web browser at the URL http://localhost:5006/myapp. Any changes you make to the plot or the widgets will be updated in real-time on the web page.





```
